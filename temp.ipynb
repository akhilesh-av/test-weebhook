```python
dffeed = pd.read_excel(r"feedback vlook.xlsx",)
dfasses = pd.read_excel(r"assessment score vlook.xlsx")
```

```python
print(df.columns)
print(dfasses.columns)
print(dffeed.columns)
```

```python
dffeed.columns = ['id', 'name', 'feedback']
```

```python
dffeed
```

```python
dfasses.columns = ['id', 'name', 'score']
```

```python
dfasses
```

```python
# Rename the 'id' column in dfasses and dffeed to avoid conflict
dfasses.rename(columns={'id': 'id_asses'}, inplace=True)
dffeed.rename(columns={'id': 'id_feed'}, inplace=True)

# Merge df with dfasses on the condition 'Emp ID' == 'id_asses' and apply suffixes to handle conflicts
df = df.merge(dfasses[['id_asses', 'score']], left_on='Emp ID', right_on='id_asses', how='left', suffixes=('', '_asses'))

# Merge df with dffeed on the condition 'Emp ID' == 'id_feed' and apply suffixes to handle conflicts
df = df.merge(dffeed[['id_feed', 'feedback']], left_on='Emp ID', right_on='id_feed', how='left', suffixes=('', '_feed'))

# Now, assign the values from 'score_asses' to 'Post Assessment Score' and 'feedback' to 'Feedback'
df['Post Assessment Score'] = df['score']
df['Feedback'] = df['feedback']

# Drop the temporary columns that were used for merging
df.drop(['id_asses', 'id_feed', 'score_asses', 'feedback'], axis=1, inplace=True)

# Check the final dataframe
print(df)
```
